```python
import matplotlib.gridspec as gridspec
def plotImageMosaic( ims, masks=None, figSize=(12,24), imCM=None, mskCM=None ):
    fig = plt.figure(figsize=figSize)
    
    nb    = len(ims)
    ncols = int(np.sqrt(nb))
    nrows = int(math.ceil(nb/ncols))

    gs = gridspec.GridSpec(nrows, ncols=int(np.sqrt(nb)), height_ratios=np.zeros(nrows)+ 2, wspace=0.0, hspace=0.0)
    
    for i in range(nb):
        inner = gridspec.GridSpecFromSubplotSpec(1 if masks is None else 2, 1,subplot_spec=gs[i], wspace=0.0, hspace=0.0)
        
        ax = plt.subplot(inner[0])
        ax.axis('off')
        ax.imshow(ims[i],cmap=imCM)
        
        if masks is not None:
            ax = plt.subplot(inner[1])
            ax.axis('off')
            ax.imshow(masks[i],cmap=mskCM)
            
```

```python
def open_image_16bit2rgb( fn ): 
    a = np.asarray(PIL.Image.open( fn ).convert("I"))
    #print(a.shape)
    #print(np.max(a))
    a = np.expand_dims(a,axis=2)
    a = np.repeat(a, 3, axis=2)
    #print(a.shape)
    return Image( pil2tensor(a, np.float32 ).div_(255) )

def getImage(): 
    return open_image_16bit2rgb(files[0])

def plots_f(rows, cols, width, height, **kwargs):
    [apply_tfms(tfms[0], get_ex(), **kwargs).show(ax=ax) for i,ax in enumerate(plt.subplots(
        rows,cols,figsize=(width,height))[1].flatten())]
```
